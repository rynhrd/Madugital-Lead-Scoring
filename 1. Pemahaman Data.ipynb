```python
import pandas as pd

# Memuat file CSV
df = pd.read_csv('data/lead_scoring.csv')

# Ringkasan statistik untuk data numerik
df_description = df.describe()

missing_values = df.isnull().sum() # Jumlah nilai yang hilang (NaN) di setiap kolom
unique_counts = df.nunique() # Jumlah nilai unik di setiap kolom
data_types = df.dtypes  # Tipe data di setiap kolom

# Menghitung persentase nilai yang hilang untuk setiap kolom
missing_percentage = (missing_values / len(df)) * 100  # Persentase nilai yang hilang
```
